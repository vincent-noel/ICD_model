```python
import maboss
import matplotlib.pyplot as plt
import time
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
file = 'ICD_latest'
```

```python
bnd = file+'.bnd'
cfg = file+'.cfg'
upp = file+'.upp'
```

```python
sim = maboss.load(bnd,cfg)
big_sim = maboss.UpdatePopulation(sim,upp)
```

# Classic simulation

```python
start = time.time()
res = big_sim.run()
middle = time.time()
print("sim. time: ", (middle-start)/60, " mins")
```

```python
middle = time.time()
table = res.get_nodes_stepwise_probability_distribution()
stop = time.time()
print("exec. time: ",(stop-middle)/60," mins")
```

```python
plt.figure()
plt.plot(table['PopRatio']*table['CD4'],label='CD4')
plt.plot(table['PopRatio']*table['CD8'],label='CD8')
plt.plot(table['PopRatio']*table['a_cDC1'],label='a_cDC1')
plt.plot(table['PopRatio']*table['a_cDC2'],label='a_cDC2')

plt.legend()
```
